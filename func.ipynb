```python
documents = [
        {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
        {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
        {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
      ]

directories = {
        '1': ['2207 876234', '11-2'],
        '2': ['10006'],
        '3': []
      }
def main():      
    while True:
        com = input(str("Введите комманду:" ))
        if com == 'p': # Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
            get_people(input('Введите номер документа:'))
        elif com == 'l': # Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам
            get_people_list() 
        elif com == 's': # Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
            get_shelf(input('Введите номер документа:'))
        elif com == 'ads': # Пункт 4. Пользователь по команде "ads" может добавить новую полку
            add_shelf(input('Введите номер полки:'))
        elif com == 'ds': # Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
            zero(input('Введите номер полки:'))    
            
        elif com == 'ad': # Пункт 1 (необязательное) Пользователь по команде "ad" может добавить новый документ в данные
            add_command(input('Введите тип документа:'), input('Введите номер документа:'), input('Введите имя:'), input('Введите номер полки (1, 2, 3):'))
        elif com == 'exit':
            break
        else:
            print('Введите правильную комманду.')

def get_people(numbers):
    i1 = []
    for i in documents:
        if numbers in i["number"]:
            i1.append(i["name"])
    if len(i1) != 0:
        print("Владелец документа:", i["name"])  
    else:
        print("Документ не найден в базе")

def get_people_list():
    for j in documents:
        for j1 in directories.items():
            if j["number"] in j1[1]:
                print(f"№:{j['number']}, тип: {(j['type'])}, владелец: {(j['name'])}, полка №:{j1[0]}")
def get_shelf(numbers):
    break_marker = False
    for s_directory in directories.items():
        for k in s_directory[1]:
            if k == numbers:
                print('Полка №', s_directory[0])
                break_marker = True
                break
        if break_marker == True:
            break
    else:
        print("Документ не найден в базе")
    
def add_command(params_type, number, name, directories_number):
    if int(directories_number) == 1 or int(directories_number) == 2 or int(directories_number) == 3:
        documents.append({"type": params_type, "number": number, "name": name})
        directories[directories_number].append(number)
    else:
        print('Введены неверные параметры!!!')
def add_shelf(qw):
    qw_list = []
    for shelf in directories.keys():
        qw_list.append(shelf)
    for shelf_1 in qw_list:
        if qw in qw_list:
            print("Такая полка уже существует.", "Текущий перечень полок:", ', '.join(map(str, qw_list)) )
            break
        else:
            qw_list.append(qw)
            print("Полка добавлена.", "Текущий перечень полок:", ', '.join(map(str, qw_list)) )
            break
def zero(ds):
    list_del = [] # общий список
    list_del_1 = [] # список без нулевых значений
    # list_del_2 = [] # список с нулевыми значениями
    for k_1, v_1 in directories.items():
        list_del.append(k_1)
        if v_1 != []:
            list_del_1.append(k_1)
    if ds in list_del and ds in list_del_1:
        print("На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:", ', '.join(map(str, list_del)))
    elif ds in list_del and ds not in list_del_1:
            print("Полка удалена. Текущий перечень полок:", ', '.join(map(str, list_del_1)))
    else:
        print("Такой полки не существует. Текущий перечень полок:", ', '.join(map(str, list_del)))  
main()
```
